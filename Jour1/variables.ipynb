# Variables et Typage Dynamique.

#### En python tout est objet, un objet est une brique, un élément de programme...

```python
'magic stèle'
```

```python
type('magic stèle')
```

Dès qu'un objet est crée il est référencé dans la mémoire de l'ordinateur. Ici cet objet est une chaîne de caractère, pour python l'objet est dit de type chaîne (string en anglais).

Les objets sont référencés dans la mémoire par une adresse compliquée.
Pour pouvoir utiliser ces objets on leur donnne un nom : la variable.
La variable ne contient pas l'objet mais sa référence.

```python
a = 13
type(a)
```

Ici la variable a pour nom "a" et elle est affectée du nombre 13, le type de l'objet référencé par "a" est du type 'int' nombre entier. 

```python
a = 13, 3.1416, 'bazard'
type(a)
```

C'est la même variable "a" mais elle référence, à présent, un nouvel objet composé de trois objets, c'est un triplet, le type de "a" a changé c'est à présent un 'tuple' (n-uplet) 

 Les noms de variables peuvent contenir des chifres, le symbole underscore '_' mais ils ne doivent **jamais** commencer par un chiffre.
Une variable peut commencer par '_' mais il vaut mieux éviter.

```python
mon_nom = 'maths'
Mon_nom = 'matheux'
## les deux variables sont différentes
print (mon_nom, Mon_nom)
```

L'exemple précédent nous montre que Python est sensible à la casse (majuscule, miniscule) les deux variables mon_nom et Mon_nom sont différentes.
Les variables doivent être **explicites** : 'score_joueur1' l'est davantage que 'scr1' ou 'x1'

Python reconnait lui même le type de la variable: on dit qu'il fait du **typage dynamique** (moins de lourdeur, nouvelle génération de logiciel de programmation) à la différence du typage statique (Java, c++ etc...)
###### Des avantages mais aussi des inconvénients...

### Les Types

```python
a = 3 + 5j
```

```python
type(a)
```

#### Observons :
Durant cette courte activité la variable "a" a changé trois fois de type. Ceci n'est pas possible avec C++, Java par exemple. Chaque variable est déclarée avec son type et dès lors on ne peut pas lui affecter des objets de type différents.
